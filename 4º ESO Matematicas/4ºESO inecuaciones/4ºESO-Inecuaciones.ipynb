## Ejercicio 5

$2x - 3 < 1 \quad  \quad  -3 \geq x + \frac {x}{2}$

Se resuelve cada inecuación por separado.
La intersección de las soluciones es la solución al sistema

$ 2x - 3 < 1 \quad  \quad  2x < 4 \quad \quad x < \frac {4}{2} \quad \quad x < 2$

Resolvemos la otra inecuación

$-3 \geq x + \frac {x}{2}$

$2 \cdot (-3) \geq 2x +x$

$-6 \geq 3x \quad \quad \frac {-6}{3} \geq x \quad \quad -2 \geq x$

También se puede expresar como $x \leq -2$

Las dos soluciones son: x < 2 y $x \leq -2$
La intersección de ambas (y por tanto, la solución del sistema) es:También

$$x \leq -2 $$
